```python
# Import the required libraries
import osmnx as ox
import geopandas as gpd
from shapely.geometry import MultiPolygon

# Define the area to fetch data for
place_name = "Banepa, Nepal"

# Fetch the building data
building_data = ox.geometries_from_place(place_name, tags={'building': True})

# Convert Polygon geometries to MultiPolygon
building_data['geometry'] = building_data['geometry'].apply(lambda x: MultiPolygon([x]) if x.geom_type == 'Polygon' else x)

# Calculate the centroid of each building
building_data['centroid'] = building_data['geometry'].centroid

# Extract latitude and longitude of the centroid
building_data['latitude'] = building_data['centroid'].apply(lambda point: point.y)
building_data['longitude'] = building_data['centroid'].apply(lambda point: point.x)

# Save the building data as a shapefile
# building_data.to_file("banepa_buildings.shp")

# Save the building data as a GeoJSON file
# building_data.to_file("banepa_buildings.geojson", driver='GeoJSON')

# Display the first few rows of the GeoDataFrame
print(building_data[['geometry', 'centroid', 'latitude', 'longitude']].head())
```

```python
# remove empty name, lat, lon from data

data = building_data.dropna(subset=['latitude', 'longitude'])
data
```

```python
data.to_csv("banepa_buildings.csv", index=False)
```
